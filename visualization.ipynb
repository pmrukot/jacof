```python
%matplotlib inline
import matplotlib.pyplot as plt

ant_systems = ['AntSystem', 'IndependentAntSystem']
benchmarks = ['attractiveness-dispersion', 'attractiveness-ratio', 'pheromone-ratio']
filenames = [system + '-' + benchmark + '.txt' for benchmark in benchmarks for system in ant_systems]

data = {}

for filename in filenames:
    with open(filename) as file:
        data[filename] = list(map(float, filter(lambda x: x!='', file.read().split('\n'))))

print(filenames)        
```

```python
iterations = list(range(1, 101))

for benchmark, i in zip(benchmarks, list(range(3))):    
    fig, ax1 = plt.subplots()
    ax1.title.set_text(benchmark)

    ax1.plot(iterations, data[filenames[i]], 'b-')
    ax1.set_xlabel('Iteration')
    ax1.set_ylabel('Classic Ant System', color='b')

    ax2 = ax1.twinx()
    ax2.plot(iterations, data[filenames[i+1]], 'r-')
    ax2.set_ylabel('Independent Ant System', color='r')

    plt.show()
```

```python
%matplotlib inline
from operator import methodcaller
from functools import reduce
import matplotlib.pyplot as plt

ant_systems = ['AntSystem', 'IndependentAntSystem']
problems = ['TravellingSalesmanProblem']
benchmarks = ['alfa-beta-gaussian', 'alfa-beta-gaussian-v2', 'alfa-beta-gaussian-v3', 'alfa-v1', 'alfa-v2', 'beta-v1', 'beta-v5', 'alfa-beta-v10']
experiment_ids = ["0", "1", "2", "3", "4"]
filenames = [system + '-' + problem + '-' + benchmark + '-' + experiment_id for problem in problems for benchmark in benchmarks for experiment_id in experiment_ids for system in ant_systems]

data = {}

for filename in filenames:
    with open(filename) as file:
        lines = list(map(methodcaller("split", ','), (filter(lambda x: x!='', file.read().split('\n')))))
        flatten_lines = reduce(lambda x,y: x+y, lines)
        current_best = list(map(float, flatten_lines[0::2]))
        global_best = list(map(float, flatten_lines[1::2]))
        data[filename] = {}
        data[filename]['current_best'] = current_best
        data[filename]['global_best'] = global_best

# print(data)
# print(filenames)
```
