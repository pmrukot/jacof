```python
%matplotlib inline
import matplotlib.pyplot as plt

ant_systems = ['AntSystem', 'IndependentAntSystem']
benchmarks = ['attractiveness-dispersion', 'attractiveness-ratio', 'pheromone-ratio']
filenames = [system + '-' + benchmark + '.txt' for benchmark in benchmarks for system in ant_systems]

data = {}

for filename in filenames:
    with open(filename) as file:
        data[filename] = list(map(float, filter(lambda x: x!='', file.read().split('\n'))))

print(filenames)        
```

```python
iterations = list(range(1, 101))

for benchmark, i in zip(benchmarks, list(range(3))):    
    fig, ax1 = plt.subplots()
    ax1.title.set_text(benchmark)

    ax1.plot(iterations, data[filenames[i]], 'b-')
    ax1.set_xlabel('Iteration')
    ax1.set_ylabel('Classic Ant System', color='b')

    ax2 = ax1.twinx()
    ax2.plot(iterations, data[filenames[i+1]], 'r-')
    ax2.set_ylabel('Independent Ant System', color='r')

    plt.show()
```
